```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('Q6.1.xlsx')
data.head()
```

```python
plt.figure(figsize=(10,6))
plt.title('Time Series of Num')
sns.lineplot(x=data['Month'],y=data['Num'],color='red')
plt.savefig('Time Series of Num.png')
plt.show()
```
